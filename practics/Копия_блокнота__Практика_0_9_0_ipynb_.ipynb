# Задание 1

Симулятор виртуального питомца

Цель: создать виртуальный симулятор домашних животных, в котором пользователи смогут заводить питомцев и ухаживать за ними.

Требования:

- Функция для усыновления питомца (имя, тип, возраст).
- Функция для того, чтобы покормить питомца, поиграть с ним или уложить его спать.
- Функция для отображения состояния питомца (голод, радость, энергия).

````python
import time


def new_pet(name, pet_type, age):
    pet = {
        "initialization": {
            "name": name,
            "type": pet_type,
            "age": age
        },
        "needs": {
            "is_starve": True,
            "is_happy": False,
            "is_exhausted": True,
            "is_sleep": False,
        },
        "time": {
            "fed_time_start": 0,
            # Количество времени, чтобы снова проголодаться (в секундах)
            "fed_time_need_passed": 10,

            "slept_time_start": 0,
            # Столько времени уйдёт на сон (в секундах)
            "slept_time_need_passed": 30
        }
    }
    return pet


def adopt_pet():
    name = input("Имя питомца: ")
    pet_type = input("Тип питомца: ")
    age = int(input("Возраст: "))

    pet = new_pet(name, pet_type, age)

    pets.append(pet)


def feed_pet(pet):
    name = pet["initialization"]["name"]
    is_starve = check_is_starve(pet)
    is_sleep = check_is_sleep(pet)

    if is_sleep:
        print(f"{name} спит")
    elif not is_sleep:
        if not is_starve:
            print(f"{name} не хочет есть")
        elif is_starve:
            pet["needs"]["is_starve"] = False
            pet["time"]["fed_time_start"] = time.time()
            print(f"{name} поел")


def check_is_starve(pet):
    is_starve = pet["needs"]["is_starve"]

    if not is_starve:
        current_time = time.time()
        fed_time_start = pet["time"]["fed_time_start"]
        fed_time_need_passed = pet["time"]["fed_time_need_passed"]

        if (current_time - fed_time_start) >= fed_time_need_passed:
            pet["needs"]["is_starve"] = True
            return True
        else:
            return False

    elif is_starve:
        return True


def play_with_pet(pet):
    name = pet["initialization"]["name"]
    is_happy = pet["needs"]["is_happy"]
    is_exhausted = pet["needs"]["is_exhausted"]
    is_sleep = check_is_sleep(pet)

    if is_sleep:
        print(f"{name} спит")
    elif not is_sleep:
        if not is_exhausted:
            pet["needs"]["is_happy"] = True
            pet["needs"]["is_exhausted"] = True
            print(f"После игры {name} чувствует себя счастливее")
        elif is_exhausted:
            print(f"{name} устал и не хочет пока что играть, возможно ему следует поспать")


def put_pet_to_bed(pet):
    name = pet["initialization"]["name"]
    is_exhausted = pet["needs"]["is_exhausted"]
    is_sleep = check_is_sleep(pet)

    if is_sleep:
        print(f"{name} уже спит")
    elif not is_sleep:

        if not is_exhausted:
            print(f"{name} пока не хочет спать, попробуйте поиграть с ним")
        elif is_exhausted:
            pet["needs"]["is_exhausted"] = False
            pet["needs"]["is_sleep"] = True

            pet["time"]["slept_time_start"] = time.time()
            print(f"{name} уснул")


def check_is_sleep(pet):
    is_sleep = pet["needs"]["is_sleep"]

    if is_sleep:
        current_time = time.time()
        slept_time_start = pet["time"]["slept_time_start"]
        slept_time_need_passed = pet["time"]["slept_time_need_passed"]
        if (current_time - slept_time_start) >= slept_time_need_passed:
            pet["needs"]["is_sleep"] = False
            return False
        else:
            return True
    elif not is_sleep:
        return False


def main():
    current_chosen_pet = None

    commands = ["show pets", "adopt", "chose",
                "feed", "put to bed", "play", "exit"]
    print(f"Список доступных комманд:\n\t{', '.join(commands)}")

    while True:
        if not pets:
            print("У вас пока нет ни одного питомца, вы можете его усыновить")
            adoption = input("Усыновить питомца (Да / Нет): ").lower()
            if adoption == "нет":
                break
            elif adoption == "да":
                adopt_pet()
            else:
                continue

        elif pets:
            pets_names = [pet["initialization"]["name"] for pet in pets]

            command = input("Что вы хотите сделать: ")

            if command == "exit":
                break

            elif command == "show pets":
                print(f"Ваши питомцы: {', '.join(pets_names)}")

            elif command == "adopt":
                adopt_pet()

            elif command == "chose":
                name = input("Введите имя питомца: ")
                try:
                    pet = pets[pets_names.index(name)]
                except ValueError:
                    print("Питомец не найден")
                else:
                    current_chosen_pet = pet

            elif command in ["feed", "put to bed", "play"]:
                if current_chosen_pet is None:
                    print("Для начала выберите питомца, с которым желаете взаимодействовать (chose)")
                    print(f"Ваши питомцы: {', '.join(pets_names)}")
                    continue
                else:
                    if command == "feed":
                        feed_pet(current_chosen_pet)
                    elif command == "put to bed":
                        put_pet_to_bed(current_chosen_pet)
                    elif command == "play":
                        play_with_pet(current_chosen_pet)

            else:
                print("Неизвестная команда")


pets = []

if __name__ == "__main__":
    main()
````

````python
import random


def new_character(name, armor, weapon, strength, intelligence, health):
    character = {
        'name': name,
        'armor': armor,
        'weapon': weapon,
        'strength': strength,
        'intelligence': intelligence,
        'health': health,
        'is_dead': False
    }
    character['weapon']['carrier'] = character
    return character


def new_weapon(name, damage_type, damage):
    return {
        'name': name,
        'damage_type': damage_type,
        'damage': damage,
        'carrier': None
    }


def new_armor(name, physic_protection, magic_resist):
    return {
        'name': name,
        'physic_protection': physic_protection,
        'magic_resist': magic_resist
    }


def calculate_damage(weapon, armor):
    max_damage = 0
    if weapon['damage_type'] == 'physical':
        max_damage = weapon['carrier']['strength'] + weapon['damage'] * \
            (1 - armor['physic_protection'] / 100)
    elif weapon['damage_type'] == 'magical':
        max_damage = weapon['carrier']['intelligence'] + weapon['damage'] * \
            (1 - armor['magic_resist'] / 100)

    total_damage = int(random.uniform(0, max_damage))
    return total_damage


def attack(character, target):
    if not character['is_dead']:
        total_damage = calculate_damage(character['weapon'], target['armor'])
        target['health'] -= total_damage

        if target['health'] > 0:
            message = (
                f'{target["name"]} получил {total_damage} урона '
                f'от {character["name"]}.\n'
                f'Оставшееся здоровье: {target["health"]}\n'
            )
        elif target['health'] <= 0:
            die(target)
            message = f'{target["name"]} пал от руки {character["name"]}.\n'
        print(message)


def die(character):
    character['is_dead'] = True


weapons = [
    {
        "name": "Кинжал",
        "damage_type": "physical",
        "damage": 8
    },
    {
        "name": "Меч",
        "damage_type": "physical",
        "damage": 15
    },
    {
        "name": "Топор",
        "damage_type": "physical",
        "damage": 20
    },
    {
        "name": "Копьё",
        "damage_type": "physical",
        "damage": 30
    },
    {
        "name": "Лук",
        "damage_type": "physical",
        "damage": 40
    },
    {
        "name": "Арбалет",
        "damage_type": "physical",
        "damage": 50
    },
    {
        "name": "Праща",
        "damage_type": "physical",
        "damage": 25
    },
    {
        "name": "Булава",
        "damage_type": "physical",
        "damage": 60
    },
    {
        "name": "Нож",
        "damage_type": "physical",
        "damage": 7
    },
    {
        "name": "Серп",
        "damage_type": "physical",
        "damage": 12
    },
    {
        "name": "Посох",
        "damage_type": "magical",
        "damage": 10
    }
]

armors = [
    {
        "name": "Кольчуга",
        "physic_protection": 20,
        "magic_resist": 15
    },
    {
        "name": "Латы",
        "physic_protection": 30,
        "magic_resist": 25
    },
    {
        "name": "Кожаная броня",
        "physic_protection": 10,
        "magic_resist": 5
    },
    {
        "name": "Доспех",
        "physic_protection": 40,
        "magic_resist": 35
    },
    {
        "name": "Кираса",
        "physic_protection": 25,
        "magic_resist": 10
    },
    {
        "name": "Шлем",
        "physic_protection": 5,
        "magic_resist": 0
    },
    {
        "name": "Щит",
        "physic_protection": 8,
        "magic_resist": 6
    },
    {
        "name": "Нагрудник",
        "physic_protection": 32,
        "magic_resist": 22
    },
    {
        "name": "Бармица",
        "physic_protection": 14,
        "magic_resist": 7
    },
    {
        "name": "Панцирь",
        "physic_protection": 45,
        "magic_resist": 37
    }
]
````

Для данной задачи используйте: https://pokeapi.co/


````python
import requests


def send_request(url, **params):
    with requests.get(url, params) as response:
        if response.ok:
            last_required_url = response.url
            data = response.json()
        else:
            response.raise_for_status()

    return data

````

````python
def add_pokemon_to_team(pokemon, team):
    if pokemon not in team:
        team.append(pokemon)
        print('Покемон добавлен в команду')
    elif pokemon in team:
        print('Этот покемон уже есть в команде')


def remove_pokemon_from_team(pokemon, team):
    if pokemon in team:
        team.remove(pokemon)
        print('Покемон удалён из команды')
    elif pokemon not in team:
        print('Этого покемона нет в команде')


def get_info_about_pokemons(team):
    for pokemon in team:
        print(pokemon['something'])


def get_pokemon_info(name):
    url = f"https://pokeapi.co/api/v2/pokemon/{name}"

    data = send_request(url)

    height = data["height"]

    weight = data["weight"]

    types = []
    for type_info in data["types"]:
        type_name = type_info["type"]["name"]
        types.append(type_name)

    abilities = []
    for ability_info in data["abilities"]:
        ability_name = ability_info["ability"]["name"]
        abilities.append(ability_name)

    return {
        "name": name,
        "height": height,
        "weight": weight,
        "abilities": abilities,
        "types": types,
    }

````
